```python
# mean
def mean(arr,n):
    s=0
    for i in range(n):
        s+=arr[i]
    return s/n



def median(arr,n):


    for i in range(n):
      for j in range(i+1, n):
        if arr[i] >= arr[j]:
            # temporary variable
            temp = arr[i]
            arr[i] = arr[j]
            arr[j] = temp

    if(n%2==0):
       return ((arr[n//2]+arr[n//2+1])/2)

    else:
        return (arr[(n+1)//2])




def mode(arr,n):
    max=0
    for i in range(n):
        ct=0
        for j in range(n):
            if(arr[i]==arr[j]):
                ct+=1

        if(ct>max):
            mode=arr[i]
            max=ct
    return (mode)

def std(arr,n):
    sum=0
    m=mean(arr,n)

    for i in range(n):
        sum+=((arr[i]-m)**2)
    sd=(sum/n)**(1/2)
    return sd

def var(arr,n):
    v=std(arr,n)**2
    return v

def skew(arr,n):
    m=mean(arr,n)
    me=median(arr,n)
    s=std(arr,n)
    sk=(3*(m-me))/s
    return sk


def kur(arr,n):
    f1=0
    m=mean(arr,n)
    s=std(arr,n)
    for i in range(n):
        f1+=(arr[i]-m)**4
    frth=f1/n
    kur=frth/s**4

    return kur
```

```python
l=[10,2,30,45,30,5,2,4,34,6,30]
n=len(l)
print("mean : ",mean(l,n))
print("median : ",median(l,n))
print("mode : ",mode(l,n))
print("standard deviation : ",std(l,n))
print("variance : ",var(l,n))
print("skewness : ",skew(l,n))
print("kurtosis : ",kur(l,n))
```
