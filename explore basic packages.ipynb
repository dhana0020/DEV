```python
import matplotlib.pyplot as plt
x=list(range(-10,10))
y=[i*2 for i in x]
plt.plot(x,y)
plt.xlabel("x   -------->")
plt.ylabel("y   ---------->")
plt.legend("line")
```

```python

import matplotlib.lines as mlines
plt.scatter(df["bill_length_mm"], df["bill_depth_mm"], c=df["species"].map({"Adelie": "blue", "Chinstrap": "red", "Gentoo": "orange"}))
plt.title("bill_length_mm vs bill_depth_mm")
plt.xlabel("bill_length_mm")
plt.ylabel("bill_depth_mm")

legend_labels = ["Adelie", "Chinstrap", "Gentoo"]
colors = ["blue", "red", "orange"]

h = [mlines.Line2D([], [], color=color, label=label,)
           for color, label in zip(colors, legend_labels)]

plt.legend(handles=h)

plt.show()
```

```python
plt.bar(df["species"],df["body_mass_g"])
plt.xlabel("species")
plt.ylabel("body_mass_g")
plt.title("species vs body_mass_g")
```

```python
plt.figure(figsize=(10,5))
plt.hist(df["bill_length_mm"],color='black',edgecolor='white')
plt.xlabel("bill_length_mm")
plt.savefig("myImage.png", format="jpeg", dpi=1080)
plt.ylabel("frequency")
plt.title("histogram of bill_length_mm")
```

```python
plt.pie(df["species"].value_counts(),labels=df["species"].unique(),autopct='%1.2f%%')
plt.title("pie chart of species")
```
