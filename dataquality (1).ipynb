```python
def report(df):


    mi = df.isnull().sum()
    mi = (mi / len(df)) * 100

    report = pd.DataFrame({
        'Feature': mi.index,
        'Missing Values': mi.values,
        'Percentage': mi.values
    })

    report = report[report['Missing Values'] > 0].sort_values(by='Percentage', ascending=False)

    print("Missing Value Report:")
    display(report)

    return report

# --- 2. Irregular Cardinality Features ---
def report2(df, threshold=10):
    """Identifies features with potentially irregular cardinality."""

    categorical_features = df.select_dtypes(include=['object']).columns
    report = {}

    for feature in categorical_features:
        unique_count = df[feature].nunique()
        if unique_count > threshold:
            report[feature] = unique_count

    print("\nIrregular Cardinality Features (Unique Count > {}):".format(threshold))
    for feature, count in report.items():
        print(f"- {feature}: {count}")

    return report

# --- 3. Outlier Summary ---
def report3(df, method='zscore', threshold=3):
    """Identifies outliers using Z-score or IQR method."""

    numerical_features = df.select_dtypes(include=['number']).columns
    report = {}

    for feature in numerical_features:
        if method == 'zscore':
            z_scores = np.abs((df[feature] - df[feature].mean()) / df[feature].std())
            outliers = df[z_scores > threshold]

        elif method == 'iqr':
            Q1 = df[feature].quantile(0.25)
            Q3 = df[feature].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
        repor[feature] = len(outliers)
    print("\nOutlier Summary ({} Method):".format(method))
    for feature, count in report.items():
        print(f"- {feature}: {count}")
    return report
report(df)
report2(df)
report3(df)
```
